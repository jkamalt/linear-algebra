#### Пример 1 <a class='anchor' id='system_p1_2_1'>
    
*Проверьте на совместность и выясните, сколько решений будет иметь система линейных уравнений:*

$$\begin{cases}
3x_1 - x_2 + x_3 = 4, \\
2x_1 - 5x_2 - 3x_3 = -17, \\
x_1 + x_2 - x_3 = 0.
\end{cases}$$

Запишем расширенную матрицу

$$\tilde{A} = \begin{pmatrix}
\left. \begin{matrix}
3 & -1 & 1 \\ 
2 & -5 & -3 \\ 
1 & 1 & -1 
\end{matrix} \right|
\begin{matrix}
4 \\ 
-17 \\
0
\end{matrix}
\end{pmatrix}$$

Столбцы исходной матрицы линейно независимы. Три линейно независимых вектора образуют в трехмерном пространстве базис и любой вектор можно выразить через линейную комбинацию базисных векторов единственным образом.

Таким образом, система совместна и имеет единственное решение.

Проверим с помощью теоремы Кронекера-Капелли. Путем элементарных преобразований приведем матрицу к ступенчатому виду.

Переставим первую и третью строки местами:

$$\begin{pmatrix}
\left. \begin{matrix}
1 & 1 & -1 \\
2 & -5 & -3 \\
3 & -1 & 1
\end{matrix} \right|
\begin{matrix}
0 \\
-17 \\
4
\end{matrix}
\end{pmatrix}$$

Выполним преобразования над строками $(2) - 2 \cdot (1)$, $(3) - 3 \cdot (1)$:

$$\begin{pmatrix}
\left. \begin{matrix}
1 & 1 & -1 \\
0 & -7 & -1 \\
0 & -4 & 4
\end{matrix} \right|
\begin{matrix}
0 \\
-17 \\
4
\end{matrix}
\end{pmatrix}$$

Третью строку разделим на 4 и поменяем местами со второй:

$$\begin{pmatrix}
\left. \begin{matrix}
1 & 1 & -1 \\
0 & -1 & 1 \\
0 & -7 & -1
\end{matrix} \right|
\begin{matrix}
0 \\
1 \\
-17
\end{matrix}
\end{pmatrix}$$

Выполним преобразование $(3) - 7 \cdot (2)$:

$$\begin{pmatrix}
\left. \begin{matrix}
1 & 1 & -1 \\
0 & -1 & 1 \\
0 & 0 & -8
\end{matrix} \right|
\begin{matrix}
0 \\
1 \\
-24
\end{matrix}
\end{pmatrix}$$

Видно, что ранг основной матрицы $A$ совпадает с рангом расширенной матрицы $\tilde{A}$ и равен числу неизвестных $n$

$$\text{rank} A = \text{rank} \tilde{A} = n = 3.$$

Тогда по теореме Кронекера-Капелли такая система совместна и определена, т.е. имеет единственное решение.

**Ответ: система совместна и имеет единственное решение.**

```python
A = np.array([[3, -1, 1], [2, -5, -3], [1, 1, -1]])
A_tilde = np.array([[3, -1, 1, 4], [2, -5, -3, -17], [1, 1, -1, 0]])

print(f'Ранг основной матрицы = {np.linalg.matrix_rank(A)}')
print(f'Ранг расширенной матрицы = {np.linalg.matrix_rank(A_tilde)}')
```

#### Пример 2 <a class='anchor' id='system_p1_2_2'>
    
*Проверьте на совместность и выясните, сколько решений будет иметь система линейных уравнений:*

$$\begin{cases}
2x_1 - 4x_2 + 6x_3 = 1, \\
x_1 - 2x_2 + 3x_3 = -2, \\
3x_1 - 6x_2 + 9x_3 = 5;
\end{cases}$$

Запишем расширенную матрицу

$$\tilde{A} = \begin{pmatrix}
\left. \begin{matrix}
2 & -4 & 6 \\ 
1 & -2 & 3 \\ 
3 & -6 & 9 
\end{matrix} \right|
\begin{matrix}
1 \\ 
-2 \\
5
\end{matrix}
\end{pmatrix}$$

Очевидно, что все столбцы исходной матрицы линейно зависимы. Это значит, что все три вектора лежат на одной прямой, то есть принадлежат некоторому одномерному линейному подпространству исходного трехмерного пространства.

Вектор правой части и столбцы исходной матрицы линейно независимы. Это значит, что вектор правой части не принадлежит линейному подпространству, то есть не лежит на этой прямой.

Таким образом, система несовместна.

Действительно, как уже показали, ранг исходной матрицы равен 1, а ранг расширенной матрицы равен 2. Покажем это также с помощью приведения матрицы к ступенчатому виду путем элементарных преобразований.

Переставим первую и вторую строки местами:

$$\begin{pmatrix}
\left. \begin{matrix}
1 & -2 & 3 \\
2 & -4 & 6 \\ 
3 & -6 & 9 
\end{matrix} \right|
\begin{matrix}
-2 \\
1 \\
5
\end{matrix}
\end{pmatrix}$$

Выполним преобразования над строками $(2) - 2 \cdot (1)$, $(3) - 3 \cdot (1)$:

$$\begin{pmatrix}
\left. \begin{matrix}
1 & -2 & 3 \\
0 & 0 & 0 \\ 
0 & 0 & 0 
\end{matrix} \right|
\begin{matrix}
-2 \\
5 \\
11
\end{matrix}
\end{pmatrix}$$

Выполнив преобразование $(3) - \frac{11}{5} \cdot (2)$ получим:

$$\begin{pmatrix}
\left. \begin{matrix}
1 & -2 & 3 \\
0 & 0 & 0
\end{matrix} \right|
\begin{matrix}
-2 \\
5
\end{matrix}
\end{pmatrix}$$

Видно, что ранг основной матрицы $A$ меньше ранга расширенной матрицы $\tilde{A}$

$$\text{rank} A = 1 < \text{rank} \tilde{A} = 2.$$

Тогда по теореме Кронекера-Капелли такая система несовместна.

**Ответ: система несовместна.**

```python
A = np.array([[2, -4, 6], [1, -2, 3], [3, -6, 9]])
A_tilde = np.array([[2, -4, 6, 1], [1, -2, 3, -2], [3, -6, 9, 5]])

print(f'Ранг основной матрицы = {np.linalg.matrix_rank(A)}')
print(f'Ранг расширенной матрицы = {np.linalg.matrix_rank(A_tilde)}')
```

#### Пример 1 <a class='anchor' id='system_p2_1_1'>
    
*Решите систему уравнений методом Крамера:*

$$\begin{cases}
x_1 - 2x_2 = 1, \\
3x_1 - 4x_2 = 7.
\end{cases}$$

Вычислим определитель матрицы

$$\det A = \begin{vmatrix}
1 & -2 \\ 
3 & -4
\end{vmatrix} = -4 + 6 = 2 \neq 0,$$

т.е. система совместна.

Вычислим определители $\det A_1$, $\det A_2$, получаемые заменой $i$-го столбца на вектор правой части.

$$\det A_1 = \begin{vmatrix}
1 & -2 \\ 
7 & -4
\end{vmatrix} = -4 + 14 = 10,$$

$$\det A_2 = \begin{vmatrix}
1 & 1 \\ 
3 & 7
\end{vmatrix} = 7 - 3 = 4.$$

Тогда по формуле Крамера получим решение:

$$x_1 = \frac{\det A_1}{\det A} = \frac{10}{2} = 5,$$

$$x_2 = \frac{\det A_2}{\det A} = \frac{4}{2} = 2.$$

**Ответ:** $\begin{pmatrix}
5 \\ 
2
\end{pmatrix}$.

```python
A = np.array([[1, -2], [3, -4]])
b = np.array([1, 7])

x = np.linalg.solve(A, b)

print(f'Матрица A =\n{A}\n')
print(f'Вектор правой части b =\n{b}\n')
print(f'Решение x_numpy =\n{x}')
```

### Задание 4 <a class='anchor' id='system_p2_4'>

*Решите систему линейных уравнений методом Холецкого:*

$$\begin{cases}
81x_1 - 45x_2 + 45x_3 = 531, \\
-45x_1 + 50x_2 - 15x_3 = -460, \\
45x_1 - 15x_2 + 38x_3 = 193.
\end{cases}$$

Матрица системы и вектор правой части имеют вид:

$$A = \begin{pmatrix}
81 & -45 & 45 \\ 
-45 & 50 & -15 \\ 
45 & -15 & 38
\end{pmatrix}, \; b = \begin{pmatrix}
531 \\ 
-460 \\ 
193
\end{pmatrix}.$$

Элементы матрицы $L$ рассчитываются по формулам:

$$l_{ii}=\sqrt{a_{ii}-\sum_{k = 1}^{i - 1}l_{ik}^{2}},$$
$$l_{ij}=\frac{1}{l_{jj}} \left( a_{ij} - \sum_{k = 1}^{j - 1}l_{ik}l_{jk} \right), \; j < i.$$


Выполним разложение $LL^T$:

$$l_{11} = \sqrt{a_{11}} = \sqrt{81} = 9,$$

$$l_{21} = \frac{a_{21}}{l_{11}} = \frac{-45}{9} = -5,$$

$$l_{31} = \frac{a_{31}}{l_{11}} = \frac{45}{9} = 5,$$

$$l_{22} = \sqrt{a_{22} - l_{21}^2} = \sqrt{50 - 25} = 5,$$

$$l_{32} = \frac{1}{l_{22}} (a_{32} - l_{31} l_{21}) = \frac{1}{5} (-15 + 25) = 2,$$

$$l_{33} = \sqrt{a_{33} - l_{31}^2 - l_{32}^2} = \sqrt{38 - 25 - 4} = 3.$$

Получили матрицу

$$L = \begin{pmatrix}
9 & 0 & 0 \\ 
-5 & 5 & 0 \\ 
5 & 2 & 3
\end{pmatrix}, \; L^T = \begin{pmatrix}
9 & -5 & 5 \\ 
0 & 5 & 2 \\ 
0 & 0 & 3
\end{pmatrix}.$$

Теперь решим систему

$$Ly = b,$$

$$\begin{cases}
9y_1 = 531, \\
-5y_1 + 5y_2 = -460, \\
5y_1 + 2y_2 + 3y_3 = 193.
\end{cases} \; \Rightarrow \; \begin{cases}
y_1 = 59, \\
y_2 = \frac{1}{5}(-460 + 5y_1) = \frac{1}{5}(-460 + 295) = -33, \\
y_3 = \frac{1}{3}(193 - 5y_1 - 2y_2) = \frac{1}{3}(193 - 295 + 66) = -12.
\end{cases}$$

Получили:

$$\begin{cases}
y_1 = 59, \\
y_2 = -33, \\
y_3 = -12.
\end{cases}$$

Затем решим систему

$$L^Tx = y,$$

$$\begin{cases}
9x_1 - 5x_2 + 5x_3 = 59, \\
5x_2 + 2x_3 = -33, \\
3x_3 = -12.
\end{cases} \; \Rightarrow \; \begin{cases}
x_3 = -4, \\
x_2 = \frac{1}{5} (-33 - 2x_3) = \frac{1}{5} (-33 + 8) = -5, \\
x_1 = \frac{1}{9} (59 + 5x_2 - 5x_3) = \frac{1}{9} (59 - 25 + 20) = 6.
\end{cases}$$

Получили решение

$$\begin{cases}
x_1 = 6, \\
x_2 = -5, \\
x_3 = -4.
\end{cases}$$

**Ответ:** $\begin{pmatrix}
6 \\ 
-5 \\
-4
\end{pmatrix}$.

```python
A = np.array([[81, -45, 45], [-45, 50, -15], [45, -15, 38]])
b = np.array([531, -460, 193])

L = np.linalg.cholesky(A)

x = np.linalg.solve(A, b)

print(f'Матрица A =\n{A}\n')
print(f'Вектор правой части b =\n{b}\n')
print(f'L_numpy =\n{L}\n')

print(f'Решение x_numpy =\n{x}')
```

### Задание 5 <a class='anchor' id='system_p2_5'>

*Напишите на Python программу с реализацией одного из изученных алгоритмов решения СЛАУ.*

Общий для всех алгоритмов метод для решения СЛАУ с треугольной матрицей, который будет использоваться при обратном ходе алгоритма.

```python
def solve_triangle(A, b, lower=True):
    """
    Решает СЛАУ с треугольной матрицей.
    A - матрица системы,
    b - правая часть системы,
    lower - если true, то матрица нижнетреугольная, иначе верхнетреугольная.
    """
    size = A.shape[0]
    indexes = range(size) if lower else reversed(range(size))
    
    x = np.zeros(size)
    for i in indexes:
        slice_indexes = slice(i) if lower else slice(size, i, -1)
        x[i] = (b[i] - A[i, slice_indexes] @ x[slice_indexes]) / A[i, i]
        
    return x
```

#### Метод Холецкого <a class='anchor' id='cholesky'>

Реализуем метод Холецкого.

```python
def solve_cholesky(A, b):
    """
    Решает СЛАУ сетодом Холецкого.
    A - матрица системы,
    b - правая часть системы.
    :return: нижняя треугольная матрица разложения L и решение СЛАУ
    """
    
    if np.linalg.det(A) == 0:
        raise Exception('Определитель матрицы равен 0')
    
    if not np.allclose(A, A.T):
        raise Exception('Матрица несимметричная')
    
    if not all(np.linalg.eigvals(A) > 0):
        raise Exception('Матрица не является положительно определенной')
    
    size = A.shape[0]
    L = np.zeros((size, size))
    for i in range(size):
        for j in range(i + 1):
            L[i, j] = np.sqrt(A[i, i] - L[i, :i] @ L[i, :i]) if i == j else (A[i, j] - L[i, :j] @ L[j, :j]) / L[j, j]
    
    y = solve_triangle(L, b)
    x = solve_triangle(L.T, y, lower=False)
    
    return L, x
```

При сравнении с решением [Задания 4](#system_p2_4) видно, что решение и полученная матрица $L$ совпадают.

```python
A = np.array([[81, -45, 45], [-45, 50, -15], [45, -15, 38]])
b = np.array([531, -460, 193])

L, x = solve_cholesky(A, b)

print(f'L =\n{L}\n')
print(f'Решение x =\n{x}')
```

#### LU - разложение <a class='anchor' id='lu'>

Реализуем $LU$ - разложение.

```python
def solve_lu(A, b):
    """
    Решает СЛАУ сетодом LU-разложения.
    A - матрица системы,
    b - правая часть системы.
    :return: нижняя треугольная и верхняя треугольная матрицы разложения L и U и решение СЛАУ
    """
    
    if np.linalg.det(A) == 0:
        raise Exception('Определитель матрицы равен 0')
    
    size = A.shape[0]
    L = np.zeros((size, size))
    U = np.zeros((size, size))

    U[0, :] = [A[0, j] for j in range(size)]
    L[:, 0] = [A[j, 0] / U[0, 0] for j in range(size)]
    
    for i in range(1, size):
        for j in range(i, size):
            U[i, j] = A[i, j] - L[i, :i] @ U[:i, j]
            L[j, i] = (A[j, i] - L[j, :i] @ U[:i, i]) / U[i, i]
    
    y = solve_triangle(L, b)    
    x = solve_triangle(U, y, lower=False)
    
    return L, U, x
```

При сравнении с решением [Задания 3](#system_p2_3) видно, что решение и полученные матрицы $L$, $U$ совпадают.

```python
A = np.array([[2, 1, 3], [11, 7, 5], [9, 8, 4]])
b = np.array([1, -6, -5])

L, U, x = solve_lu(A, b)

print(f'Матрица A =\n{A}\n')
print(f'LU =\n{np.dot(L, U)}\n')

print(f'L =\n{L}\n')
print(f'U =\n{U}\n')

print(f'Решение x =\n{x}')
```
