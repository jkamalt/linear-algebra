# Линейные преобразования <a class='anchor' id='lin_map'>

### Задание 1 <a class='anchor' id='lin_map_1'>

*Найдите собственные векторы и собственные значения для линейного оператора, заданного матрицей:*

$$A = \begin{pmatrix}
-1 & -6 \\ 
2 & 6
\end{pmatrix}.$$

Составим характеристическое уравнение

$$|A - \lambda I| = 0,$$

$$\begin{vmatrix}
-1 - \lambda & -6 \\ 
2 & 6 - \lambda
\end{vmatrix} = 0,$$

$$(-1 - \lambda)(6 - \lambda) + 12 = 0,$$

$$-6 + \lambda - 6\lambda + \lambda^2 + 12 = 0,$$

$$\lambda^2 - 5\lambda + 6 = 0.$$

Корни характеристического уравнения:

$$\lambda_1 = 2, \lambda_2 = 3.$$

Нашли собственные значения оператора. Найдем собственные векторы, подставив найденные собственные значения в выражение

$$(A - \lambda I)x = 0.$$

1. $\lambda_1 = 2$

$$\begin{pmatrix}
-3 & -6 \\ 
2 & 4
\end{pmatrix}
\begin{pmatrix}
x_1 \\ 
x_2
\end{pmatrix} = 0 \Rightarrow
\begin{pmatrix}
1 & 2 \\ 
1 & 2
\end{pmatrix}
\begin{pmatrix}
x_1 \\ 
x_2
\end{pmatrix}= 0.$$

Второе уравнение совпадает с первым, поэтому его можно убрать. Получим уравнение

$$x_1 + 2x_2 = 0.$$

Выразим $x_1$ через $x_2$:

$$x_1 = -2x_2,$$

где $x_2$ может принимать любое значение.

Тогда векторы вида

$$\begin{pmatrix}
x_1 \\ 
x_2
\end{pmatrix} = \begin{pmatrix}
-2C \\ 
C
\end{pmatrix},$$

где $C$ - произвольная константа, являются собственными векторами для оператора $A$.

2. $\lambda_2 = 3$

$$\begin{pmatrix}
-4 & -6 \\ 
2 & 3
\end{pmatrix}
\begin{pmatrix}
x_1 \\ 
x_2
\end{pmatrix} = 0 \Rightarrow
\begin{pmatrix}
2 & 3 \\ 
2 & 3
\end{pmatrix}
\begin{pmatrix}
x_1 \\ 
x_2
\end{pmatrix}= 0.$$

Второе уравнение совпадает с первым, поэтому его можно убрать. Получим уравнение

$$2x_1 + 3x_2 = 0.$$

Выразим $x_1$ через $x_2$:

$$x_1 = -\frac{3}{2}x_2,$$

где $x_2$ может принимать любое значение.

Тогда векторы вида

$$\begin{pmatrix}
x_1 \\ 
x_2
\end{pmatrix} = \begin{pmatrix}
-\frac{3}{2}C \\ 
C
\end{pmatrix},$$

где $C$ - произвольная константа, являются собственными векторами для оператора $A$.

**Ответ: $\lambda_1 = 2$, $\lambda_2 = 3$ - собственные значения, $\begin{pmatrix}
-2C_1 \\ 
C_1
\end{pmatrix}$, $\begin{pmatrix}
-\frac{3}{2}C_2 \\ 
C_2
\end{pmatrix}$ - собственные векторы, где $C_1$, $C_2$ - произвольные константы.**

### Задание 2 <a class='anchor' id='lin_map_2'>

*Дан оператор поворота на 180 градусов, задаваемый матрицей:*

$$A = \begin{pmatrix}
-1 & 0 \\ 
0 & -1
\end{pmatrix}.$$

*Покажите, что __любой__ вектор считается для него собственным.*

Если вектор $x = (x_1, x_2)$ является собственным вектором линейного оператора, то должно существовать вещественно число $\lambda$ такое, что

$$(A - \lambda I)x = 0.$$

Подставим значения

$$\begin{pmatrix}
-1 - \lambda & 0 \\ 
0 & -1 - \lambda
\end{pmatrix} 
\begin{pmatrix}
x_1 \\ 
x_2
\end{pmatrix}
= 0.$$

$$\begin{cases}
(-1 - \lambda) x_1 = 0 \\ 
(-1 - \lambda) x_2 = 0
\end{cases} \Rightarrow
\lambda = -1.$$

Т.е. любому вектору $x = (x_1, x_2)$ соответствует собственное значение $\lambda = -1$. Это очевидно, т.к. оператор поворота на 180 градусов переводит любой вектор в противоположно направленный вектор такой же длины.

Таким образом, любой вектор является собственным вектором линейного оператора поворота на 180 градусов.

**Ответ: любой вектор является собственным вектором с собственным значением $\lambda = -1$.**

### Задание 3 <a class='anchor' id='lin_map_3'>

*Пусть линейный оператор задан матрицей:*

$$A = \begin{pmatrix}
1 & 1 \\ 
-1 & 3
\end{pmatrix}.$$

*Установите, считается ли вектор $x = (1, 1)$ собственным вектором этого линейного оператора.*

Если заданный вектор является собственным вектором линейного оператора, то должно существовать вещественно число $\lambda$ такое, что

$$(A - \lambda I)x = 0.$$

Подставим значения

$$\begin{pmatrix}
1 - \lambda & 1 \\ 
-1 & 3 - \lambda
\end{pmatrix} 
\begin{pmatrix}
1 \\ 
1
\end{pmatrix}
= 0.$$

$$\begin{cases}
1 - \lambda + 1 = 0 \\ 
-1 + 3 - \lambda = 0
\end{cases} \Rightarrow
\lambda = 2.$$

Таким образом, вектор $x = (1, 1)$ является собственным вектором линейного оператора $A$, соответствующий собственному значению $\lambda = 2$.

**Ответ: является собственным вектором.**

### Задание 4 <a class='anchor' id='lin_map_4'>

*Пусть линейный оператор задан матрицей:*

$$A = \begin{pmatrix}
0 & 3 & 0 \\
3 & 0 & 0 \\
0 & 0 & 3
\end{pmatrix}.$$

*Установите, считается ли вектор $x = (3, -3, -4)$ собственным вектором этого линейного оператора.*

Если заданный вектор является собственным вектором линейного оператора, то должно существовать вещественно число $\lambda$ такое, что

$$(A - \lambda I)x = 0.$$

Подставим значения

$$\begin{pmatrix}
-\lambda & 3 & 0 \\ 
3 & -\lambda & 0 \\
0 & 0 & 3 - \lambda
\end{pmatrix} 
\begin{pmatrix}
3 \\ 
-3 \\
-4
\end{pmatrix}
= 0.$$

$$\begin{cases}
-3\lambda -9 = 0 \\ 
9 + 3\lambda = 0 \\
-12 + 4\lambda = 0
\end{cases} \Rightarrow
\begin{cases}
\lambda = -3 \\
\lambda = -3 \\
\lambda = 3
\end{cases}.$$

Такая система не имеет смысла. Таким образом, вектор $x = (3, -3, -4)$ не является собственным вектором линейного оператора $A$.

**Ответ: не является собственным вектором.**
