### Задание 5 <a class='anchor' id='lin_space_p1_5'>

*Установите, считается ли линейным подпространством:*

*а) совокупность всех векторов трёхмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;*
    
*б) все векторы, считающиеся линейными комбинациями данных векторов $\{u_{1}, u_{2} \ldots, u_{n}\}$.*

Подмножество $L$ линейного пространства $V$ считается его подпространством только тогда, когда для любых элементов $u, v\in L$ и любого $\alpha \in \mathbb{R}$ выполняются условия:

$$1)\ u + v \in L;$$

$$2)\ \alpha \cdot u \in L.$$

Для заданных множеств проверим указанные условия.

#### Пример 1 <a class='anchor' id='lin_space_p1_5_1'>
    
*Считается ли линейным подпространством совокупность всех векторов трёхмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю.*

Векторы $L$, у которых по крайней мере одна из первых двух координат равна нулю - это векторы вида

$$(0, y, z),$$

$$(x, 0, z),$$

$$(0, 0, z).$$

Первое условие не выполняется. Возьмем векторы $u = (0, y, z)$, $v = (x, 0, z) \in L$, тогда их сумма:

$$u + v = (0, y, z) + (x, 0, z) = (x, y, z) \notin L.$$

Т.е. заданное подмножество не является линейным подпространством.

**Ответ: не является линейным подпространством.**

#### Пример 2 <a class='anchor' id='lin_space_p1_5_2'>

*Считается ли линейным подпространством все векторы, считающиеся линейными комбинациями данных векторов $\{u_{1}, u_{2} \ldots, u_{n}\}$.*

Векторы $L$, считающиеся линейными комбинациями данных векторов $\{u_{1}, u_{2} \ldots, u_{n}\}$ - это векторы вида

$$\sum_{i = 1}^n \lambda_i u_i,$$

где $\lambda_i$ одновременно не равны 0.

Очевидно, что условия выполняются. Возьмем векторы $u = \sum_{i = 1}^n \lambda_i u_i$, $v = \sum_{i = 1}^n \alpha_i u_i \in L$. Тогда:

$$u + v = \sum_{i = 1}^n \lambda_i u_i + \sum_{i = 1}^n \alpha_i u_i = \sum_{i = 1}^n (\lambda_i + \alpha_i) u_i \in L,$$

$$\alpha \cdot u = \alpha \sum_{i = 1}^n \lambda_i u_i = \sum_{i = 1}^n (\alpha \lambda_i) u_i \in L.$$

Т.е. заданное подмножество является линейным подпространством.

**Ответ: является линейным подпространством.**

Скалярное произведение векторов $x, y \in \mathbb{R^n}$ вычисляется по формуле

$$(x, y) = \sum_{i = 1}^n x_i y_i.$$

а) $x = (0,-3, 6), y = (-4, 7, 9).$

Скалярное произведение равно

$$(x, y) = 0 \cdot (-4) + (-3) \cdot 7 + 6 \cdot 9 = -21 + 54 = 33.$$

**Ответ: 33.**

б) $x = (7, -4, 0, 1), y = (-3, 1, 11, 2).$

Скалярное произведение равно

$$(x, y) = 7 \cdot (-3) + (-4) \cdot 1 + 0 \cdot 11 + 1 \cdot 2 = -21 - 4 + 2 = -23.$$

**Ответ: -23.**

```python
# Проверка
x = np.array([0,-3, 6])
y = np.array([-4, 7, 9])
print(f'а) (x, y) = {np.dot(x, y)}')

x = np.array([7, -4, 0, 1])
y = np.array([-3, 1, 11, 2])
print(f'б) (x, y) = {np.dot(x, y)}')
```

### Задание 2 <a class='anchor' id='lin_space_p2_2'>

*Найдите нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.*

Норма вектора $x$ в нормированном евклидовом пространстве вычисляется по формуле

$$\left\|x\right\| = \sqrt{(x, x)}.$$

Косинус угла между двумя произвольными элементами $x, y$ в евклидовом пространстве определяется как

$$\cos \varphi = \frac{(x, y)}{\left\|x\right\| \cdot \left\|y\right\|}.$$

Вычислим нормы векторов $x, y \in \mathbb{R^n}$:

$$\left\|x\right\| = \sqrt{(x, x)} = \sqrt{\sum_{i = 1}^n x_i^2} = \sqrt{4^2 + 2^2 + 4^2} = \sqrt{36} = 6,$$

$$\left\|y\right\| = \sqrt{(y, y)} = \sqrt{\sum_{i = 1}^n y_i^2} = \sqrt{12^2 + 3^2 + 4^2} = \sqrt{169} = 13.$$

Вычислим скалярное произведение этих векторов:

$$(x, y) = \sum_{i = 1}^n x_i y_i = 4 \cdot 12 + 2 \cdot 3 + 4 \cdot 4 = 48 + 6 + 16 = 70.$$

Тогда косинус угла между ними равен

$$\cos \varphi = \frac{(x, y)}{\left\|x\right\| \cdot \left\|y\right\|} = \frac{70}{6 \cdot 13} = \frac{70}{78} \approx 0.897.$$

Угол равен

$$\varphi = \arccos(0.897) \approx 0.457 \; \text{рад} \approx 26 \; \text{град}.$$

**Ответ:** $\left\|x\right\| = 6, \left\|y\right\| = 13, \varphi \approx 26 \; \text{град}.$

```python
# Проверка
x = np.array([4,2,4])
y = np.array([12,3,4])

norm_x = np.linalg.norm(x, ord=2)
norm_y = np.linalg.norm(y, ord=2)
dot_prod = np.dot(x, y)
angle_rad = np.arccos(dot_prod / (norm_x * norm_y))

print(f'||x|| = {norm_x}')
print(f'||y|| = {norm_y}')
print(f'(x, y) = {dot_prod}')
print(f'phi = {angle_rad} rad = {np.rad2deg(angle_rad)} grad')
```

### Задание 3 <a class='anchor' id='lin_space_p2_3'>

*Определите, будет ли линейное пространство евклидовым, если за скалярное произведение принять:*

*а) произведение длин векторов;*
    
*б) утроенное обычное скалярное произведение векторов.*

Линейное пространство над полем вещественных чисел называется _евклидовым пространством_, если на нём введено правило, согласно которому каждой упорядоченной паре векторов $x$ и $y$ поставлено в соответствие вещественное число, называемое _скалярным произведением_ и обозначаемое символом $(x, y)$.

Указанное правило должно подчиняться четырём аксиомам:

1) $(x, y) = (y, x).$

2) $(\lambda x, y)=\lambda(x, y).$

3) $(x_1 + x_2, y)=(x_1, y) + (x_2, y).$

4) $(x, x) \geq 0$, причём $(x, x) = 0 \Leftrightarrow x = 0.$

#### Пример 1 <a class='anchor' id='lin_space_p2_3_1'>
    
*Будет ли линейное пространство евклидовым, если за скалярное произведение принять произведение длин векторов.*

Предположим

$$(x, y) = \left\|x\right\| \cdot \left\|y\right\|.$$
    
Проверим аксиомы скалярного произведения

1) $(x, y) = \left\|x\right\| \cdot \left\|y\right\| = \left\|y\right\| \cdot \left\|x\right\| = (y, x)$,

2) $(\lambda x, y) = \left\|\lambda x\right\| \cdot \left\|y\right\| = |\lambda| \left\|x\right\| \cdot \left\|y\right\| = |\lambda| \; (x, y) \neq \lambda (x, y)$,

3) $(x_1 + x_2, y) = \left\|x_1 + x_2\right\| \cdot \left\|y\right\| \leq (\left\|x_1\right\| + \left\|x_2\right\|) \cdot \left\|y\right\| = (x_1, y) + (x_2, y)$,

4) $(x, x) = \left\|x\right\|^2 \geq 0$, причем $\left\|x\right\|^2 = 0 \Leftrightarrow x = 0$.

Вторая и третья аксиомы не выполняются, значит линейное пространство со скалярным произведением, заданным как произведение длин векторов не будет евклидовым.

**Ответ: не является евклидовым.**

#### Пример 2 <a class='anchor' id='lin_space_p2_3_2'>
    
*Будет ли линейное пространство евклидовым, если за скалярное произведение принять утроенное обычное скалярное произведение векторов.*

Предположим

$$(x, y) = 3 \langle x, y \rangle,$$

где для $\langle x, y \rangle$ - обычное скалярное произведение, удовлетворяющее аксиомам скалярного произведения.
    
Проверим аксиомы скалярного произведения

1) $(x, y) = 3\langle x, y \rangle = 3\langle y, x \rangle = (y, x)$,

2) $(\lambda x, y) = 3\langle \lambda x, y \rangle = 3 \lambda \langle x, y \rangle = \lambda (x, y)$,

3) $(x_1 + x_2, y) = 3\langle x_1 + x_2, y \rangle = 3\langle x_1, y \rangle + 3\langle x_2, y \rangle = (x_1, y) + (x_2, y)$,

4) $(x, x) = 3\langle x, x \rangle \geq 0$, причем $3\langle x, x \rangle = 0 \Leftrightarrow x = 0$.

Аксиомы выполняются, значит линейное пространство со скалярным произведением, заданным как утроенное обычное скалярное произведение векторов будет евклидовым.

**Ответ: является евклидовым.**

### Задание 4 <a class='anchor' id='lin_space_p2_4'>

*Выясните, какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$:*

*а) $(1, 0, 0), (0, 0, 1);$*

*б) $\left( \frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}}, 0 \right), \left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}, 0 \right), (0, 0, 1);$*

*в) $\left( \frac{1}{2}, -\frac{1}{2}, 0 \right), \left( 0, \frac{1}{2}, \frac{1}{2} \right), (0, 0, 1);$*

*г) $(1, 0, 0), (0, 1, 0), (0, 0, 1)?$*

В конечномерном евклидовом пространстве базис $\{e_1, e_2 \ldots, e_n\}$ называется ортонормированным, если:

$$(e_i, e_j) = 0 \; \forall i \neq j, (e_i, e_i) = 1 \; \forall i \in [1, n].$$

#### Пример 1 <a class='anchor' id='lin_space_p2_4_1'>
    
*Векторы $e_1 = (1, 0, 0), e_2 = (0, 0, 1)$.*

Проверим, что векторы ортогональны друг другу

$$(e_1, e_2) = 1 \cdot 0 + 0 \cdot 0 + 0 \cdot 1 = 0.$$

Проверим, что векторы имеют единичную длину

$$(e_1, e_1) = 1 + 0 + 0 = 1,$$

$$(e_2, e_2) = 0 + 0 + 1 = 1.$$

Условия ортонормированности выполняются, однако базис трехмерного пространства $\mathbb{R}^{3}$ не может состоять из двух векторов.

**Ответ: не образуют.**

#### Пример 2 <a class='anchor' id='lin_space_p2_4_2'>
    
*Векторы $e_1 = \left( \frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}}, 0 \right), e_2 = \left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}, 0 \right), e_3 = (0, 0, 1)$.*

Проверим, что векторы ортогональны друг другу

$$(e_1, e_2) = \frac{1}{\sqrt{2}} \cdot \frac{1}{\sqrt{2}} + \left(-\frac{1}{\sqrt{2}}\right) \cdot \frac{1}{\sqrt{2}} + 0 \cdot 0 = 0,$$

$$(e_1, e_3) = \frac{1}{\sqrt{2}} \cdot 0 + \left(-\frac{1}{\sqrt{2}}\right) \cdot 0 + 0 \cdot 1 = 0,$$
    
$$(e_2, e_3) = \frac{1}{\sqrt{2}} \cdot 0 + \frac{1}{\sqrt{2}} \cdot 0 + 0 \cdot 1 = 0.$$

Проверим, что векторы имеют единичную длину

$$(e_1, e_1) = \frac{1}{2} + \frac{1}{2} + 0 = 1,$$

$$(e_2, e_2) = \frac{1}{2} + \frac{1}{2} + 0 = 1,$$

$$(e_3, e_3) = 0 + 0 + 1 = 1.$$

Условия определения ортонормированного базиса выполняются.

**Ответ: образуют.**

#### Пример 3 <a class='anchor' id='lin_space_p2_4_3'>
    
*Векторы $e_1 = \left( \frac{1}{2}, -\frac{1}{2}, 0 \right), e_2 = \left( 0, \frac{1}{2}, \frac{1}{2} \right), e_3 = (0, 0, 1)$.*

Проверим, что векторы ортогональны друг другу

$$(e_1, e_2) = \frac{1}{2} \cdot 0 + \left(-\frac{1}{2}\right) \cdot \frac{1}{2} + 0 \cdot \frac{1}{2} = -\frac{1}{4},$$

$$(e_1, e_3) = \frac{1}{2} \cdot 0 + \left(-\frac{1}{2}\right) \cdot 0 + 0 \cdot 1 = 0,$$

Не все векторы ортогональный друг другу. Первое условие определения ортонормированного базиса не выполняется.

**Ответ: не образуют.**

#### Пример 4 <a class='anchor' id='lin_space_p2_4_4'>
